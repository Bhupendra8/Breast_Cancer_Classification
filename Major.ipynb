## Normalization using MinmaxScaler and StandardScaler 

```python
def scale_numeric_columns(df, method='normalize'):
    df_scaled = df.copy()

    numeric_columns = df_scaled.select_dtypes(include=['float64', 'float32', 'int64', 'int32']).columns
    
    if method == 'normalize':
        scaler = MinMaxScaler()
    elif method == 'standardize':
        scaler = StandardScaler()
    else:
        raise ValueError("Invalid method. Use 'normalize' or 'standardize'.")
    
    df_scaled[numeric_columns] = scaler.fit_transform(df_scaled[numeric_columns])
    
    return df_scaled
```

## LIME Implementation 

### Ranking the features by applying LIME on all instances and then aggregating them

```python
# !pip install lime
import warnings
warnings.filterwarnings('ignore')
from lime import lime_tabular
from sklearn.ensemble import RandomForestClassifier
import numpy as np


rf2 = RandomForestClassifier(**best_params)
rf2.fit(X,Y.values.ravel())

explainer = lime_tabular.LimeTabularExplainer(X.values, 
                                              feature_names=X.columns.tolist(), 
                                              class_names=['Benign', 'Malignant'], 
                                              discretize_continuous=True)

# Choose instances from the test set to explain
instances = []
instances.append(X.iloc[0:564])

feature_importance_sum = np.zeros(len(X.columns))

# Get a LIME explanation for each instance
num_instances = 0
for instance_to_explain in instances:
    for _, part_inst in instance_to_explain.iterrows():
        exp = explainer.explain_instance(part_inst, rf2.predict_proba, num_features=len(X.columns))
        feature_importance = np.array([imp for _, imp in exp.as_list()])
        feature_importance_sum += feature_importance
        num_instances += 1

# Calculate the average importance for each feature
average_importance = feature_importance_sum / num_instances

# Sort features based on average importance
sorted_features_indices = np.argsort(average_importance)[::-1]
sorted_features = X.columns[sorted_features_indices]

# Print the most important features according to their average importance
print("Most important features according to average importance:")

for feature in sorted_features:
    print(feature)
```
